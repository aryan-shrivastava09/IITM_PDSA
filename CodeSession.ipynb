# Check Distinct

```python
def merge(A,B):
    m,n = len(A), len(B)
    C,i,j,k = [],0,0,0
    while k<i+j:
        if i==m:
            C.extend(B[j:])
            k+=(n-j)
        elif j==n:
            C.extend(A[i:])
            k+=(m-i)
        elif A[i]<B[j]:
            C.append(A[i])
            i,k=i+1,k+1
        else:
            C.append(B[j])
            j,k=i+1,k+1
    return C

def mergesort(l):
    n = len(l)
    if n<=1:
        return l
    L = mergesort(l[:n//2])
    R = mergesort(l[n//2:])
    l1 = merge(L,R)
    return l1
```

```python
def check_distinct(L):
    L = mergesort(L)
    for i in range(len(L)-1):
        if L[i] == L[i+1]:
            return 0
    return 1
```

```python
def create_movie_theatre(n):
    H = []
    for i in range(n):
        H.append([])
    return H

def put_element(x,H):
    H[x%100].append(x)
    return H

def put_list(L,H):
    for x in L:
        H= put_element(x,H)
    return H
```

```python
def present_or_not(x,V):
    flag = 0
    for i in range(len(V)):
        if x == V[i]:
            flag = 1
    return flag
```

```python
def check_distinct_optimized(L):
    H = create_movie_theatre(len(L))
    flag =1
    for i in range(len[L]):
        seat_number =L[i]%len(L)
        if present_or_not(L[i],H[seat_number]):
            flag = 0
        if flag==1:
            H[seat_number].append(L[i])
    return flag
```
